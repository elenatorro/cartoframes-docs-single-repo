# Download a Dataset from CARTO

This example illustrates how to download a Dataset from CARTO and convert it to a DataFrame

> Note: CARTO Account credentials are needed to reproduce this example. https://carto.com/signup

```python
from cartoframes.auth import set_default_credentials

set_default_credentials('cartoframes')
```

```python
from cartoframes.data import Dataset

ds = Dataset('global_power_plants')
df = ds.download()
df.head()
```
