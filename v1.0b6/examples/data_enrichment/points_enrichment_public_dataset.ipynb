```python
import geopandas as gpd
from cartoframes.auth import set_default_credentials

set_default_credentials()
```

```python
arlington_file = '../files/bikes.geojson'
bikeshare_df = gpd.read_file(arlington_file)
bikeshare_df.head(10)
```

```python
from cartoframes.data.observatory import Dataset, Catalog

dataset = Dataset.get('carto-do-public-data.usa_acs.demographics_acs_usa_censustractclipped_2015_5yrs_20132017')
variables = dataset.variables
variables
```

```python
variable = variables.get('no_cars_d19dfd10')
variable
```

```python
from cartoframes.data.observatory import Enrichment

enrichment = Enrichment()

enriched_dataset_cdf = enrichment.enrich_points(
    bikeshare_df,
    variables=['no_cars_d19dfd10']
)
```

```python
enriched_dataset_cdf.head(10)
```

```python
from cartoframes.viz import Map
from cartoframes.viz.helpers import size_continuous_layer

Map(size_continuous_layer(enriched_dataset_cdf, value='no_cars'))
```
