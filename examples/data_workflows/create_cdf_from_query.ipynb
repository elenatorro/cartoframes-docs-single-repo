# Create a CartoDataFrame from a SQL Query

This example illustrates how to create a CartoDataFrame from CARTO using a SQL Query

> Note: CARTO Account credentials are needed to reproduce this example. https://carto.com/signup

```python
from cartoframes.auth import set_default_credentials

set_default_credentials('cartoframes')
```

```python
from cartoframes import CartoDataFrame

cdf = CartoDataFrame.from_carto("SELECT * FROM global_power_plants WHERE country IN ('Brazil')")
cdf.head()
```
