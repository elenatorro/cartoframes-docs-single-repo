# Paris Remarkable Trees from 1700 to 2000

```python
import pandas
from pandas.io.json import json_normalize
import json

local_file_path = '../files/arbres_remarquables_paris.json'
with open(local_file_path, 'r') as json_file:
    data_json = json.load(json_file)

df = pandas.DataFrame(json_normalize(data_json['records']))
df.head()
```

```python
df.columns = df.columns.str.replace(".", "_")
df[['lon','lat']] = pandas.DataFrame(df['geometry_coordinates'].values.tolist(), index=df.index)
df.loc[:, ['geometry_coordinates', 'lon', 'lat']]
```

```python
from cartoframes import CartoDataFrame

cdf = CartoDataFrame(df).set_geometry_from_xy('lon', 'lat')
cdf.head()
```

```python
cdf.viz()
```

```python
cdf['plantation_date'] = pandas.to_datetime(cdf['fields_dateplantation'], errors='coerce')
cdf = cdf[cdf['plantation_date'].notnull()]
cdf['plantation_year'] = pandas.to_numeric(cdf['plantation_date'].dt.strftime('%Y'))

cdf.loc[:, ['plantation_date', 'plantation_year']]
```

```python
from cartoframes.viz.helpers import color_bins_layer

color_bins_layer(cdf, 'plantation_year', breaks=[1700, 1800, 1900, 2000], palette='reverse(Mint)')
```
